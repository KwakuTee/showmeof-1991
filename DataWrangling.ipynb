```python
landing_outcomes=df["Outcome"].value_counts()
landing_outcomes
```

```python
landing=pd.DataFrame(landing_outcomes)
landing.reset_index()
```

```python
for i, outcomes in enumerate(landing_outcomes.keys()):
    print(i, outcomes)
    
```

```python
bad_outcomes=landing_outcomes.keys()[[1, 3, 5, 7 ]]
bad_outcomes
```

```python
landing_class=[]
for x in df["Outcome"]:
    if x in bad_outcomes:
        landing_class.append(0)
    else:
        landing_class.append(1)

print(landing_class)
```

```python
df["Class"]=landing_class
df.head(10)
```

```python
df["Class"].mean()
```
